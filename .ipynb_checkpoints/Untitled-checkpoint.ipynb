```python
# CREATE DATAFRAME COMBINING THE 2 PREVIOUS DATAFRAMES
# IN ORDER TO MAKE A JANUARY - DATAFRAME

january = pd.concat([dfs_january1, dfs_january2])
january.shape
```

```python
# Using the ExcelFile class to read multiple sheets (Excel file is only passed once)

# Better performance than read_excel (there, the Excel file is passed every time, hence the file is loaded again for every sheet)

xlsx = pd.ExcelFile("prueba.xlsx")
parts_sheets = []
for sheet in xlsx.sheet_names:
    parts_sheets.append(xlsx.parse(sheet))
all_df = pd.concat(parts_sheets)

all_df.reset_index(drop=True, inplace=True) # reset the values of 

len(all_df) # 3 files = 50 + 75 + 20 = 145
print(all_df)
```

```python
len(january)
```

```python
# it will take several minutes to process

xlsx_all = pd.ExcelFile("2017_AQ.xlsx")
months_sheets = []
for sheet in xlsx_all.sheet_names:
    months_sheets.append(xlsx_all.parse(sheet))
all_data = pd.concat(months_sheets)

len(all_data) # 3 files = 100.000 * 7 + 99848
```

```python
sorted_by_PM10 = all_data.sort_values(['PM10'], ascending=False)
sorted_by_PM10["PM10"].head(50)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

sorted_by_PM10["PM10"].head(10).plot(kind="barh")
plt.show()
```

```python
# Hallazgo enigmático de ordenar PM10 y plotear la velocidad del viento

# Cómo es el viento en las zonas más contaminadas?

sorted_by_PM10["WINDSPEED"].head(40).plot(kind="barh")
plt.show()
```

```python
# plotear PM10, WINDSPEED y IDSTATION

col = ['coral', 'blue', 'black', 'm']
col2 = ['0','1']

# plot
fig = plt.figure(figsize = (25,10))
ax = fig.subplots()

for c, category in zip(col, col2):
    xs = all_df[all_df['PM10'] == category].T[0]
    ys = all_df[all_df['PM10'] == category].T[1]
   
    ax.scatter(xs, ys, c = c, marker='o')

ax.grid(color='gray', linestyle=':', linewidth=2, alpha=0.2)
ax.set_xlabel('\nX Label')
ax.set_ylabel('\nY Label')

plt.show()
```
